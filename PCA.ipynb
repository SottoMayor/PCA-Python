```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Lendo o arquivo csv
df = pd.read_csv('england_premier_league_table_22.csv', delimiter=';')

display(df)

# Separando o nome dos times
squad = df['Squad']
df = df.drop(columns=['Rk', 'Squad', 'Top Team Scorer'])

# Definindo X como as variáveis restantes
X = df.values

# Criando o objeto PCA e definindo 2 componentes
pca = PCA(n_components=2)

# Aplicando o PCA nas variáveis X
X_pca = pca.fit_transform(X)

# Imprimindo os valores das duas componentes principais para cada time
for i, name in enumerate(squad):
    print(name, X_pca[i])

    
# Plotando um gráfico de dispersão com base na colocação geral

# Adicionando o nome de cada time no gráfico
for i, name in enumerate(squad):
    plt.text(X_pca[i, 0], X_pca[i, 1], f'{i + 1}°: {name}')

plt.scatter(X_pca[:, 0], X_pca[:, 1])
plt.title('Gráfico de Dispersão da Colocação Geral')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.show()


# Plotando um gráfico de dispersão com base em gols feitos e sofridos

# Seleciona as colunas numéricas para análise de PCA
data = df
num_cols = ['MP', 'W', 'D', 'L', 'GF', 'GA', 'GD', 'Pts', 'Pts;MP', 'xG', 'xGA', 'xGD', 'xGD;90', 'Attendance']

# Normalizando os dados
scaler = StandardScaler()
data[num_cols] = scaler.fit_transform(data[num_cols])

# Aplicando a PCA com 2 componentes
pca = PCA(n_components=2)
principal_components = pca.fit_transform(data[num_cols])

# Criando um novo DataFrame com as componentes principais
pc_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])


# Define as cores para os pontos do gráfico
colors = df['GF'].values

# Define as legendas para os pontos do gráfico
labels = df.index

# Plota o gráfico de dispersão com as duas primeiras componentes principais
fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(principal_components[:, 0], principal_components[:, 1], c=colors, cmap='cool', alpha=0.8)
    
# Adicionando o nome de cada time no gráfico
for i, name in enumerate(squad):
    plt.text(principal_components[i, 0], principal_components[i, 1], f'{name}')

# Adiciona as legendas aos eixos
ax.set_xlabel('Primeira componente principal')
ax.set_ylabel('Segunda componente principal')

plt.title('Gráfico de dispersão por gols marcados e sofridos')
plt.show()
```
